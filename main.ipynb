KOLKATA CARS 

```python
import pandas as pd
import ast

# Load the Excel file into a pandas DataFrame
df = pd.read_excel('kolkata_cars.xlsx')

# Columns we are working with
columns_to_extract = ['new_car_detail', 'new_car_overview', 'new_car_feature', 'new_car_specs']

flattened_car_data = []

# Iterate through each row
for index, row in df.iterrows():
    car_info = {}

    # Process each of the columns
    for column in columns_to_extract:
        column_data = row[column]
        
        # Convert string to dictionary if needed
        if isinstance(column_data, str):
            try:
                column_data = ast.literal_eval(column_data)
            except:
                column_data = {}

        # If the column data is a dictionary, process it
        if isinstance(column_data, dict):
            if column == 'new_car_detail':
                car_info.update({
                    "Ignition Type": column_data.get("it"),
                    "Fuel Type": column_data.get("ft"),
                    "Body Type": column_data.get("bt"),
                    "Kilometers Driven": column_data.get("km"),
                    "Transmission Type": column_data.get("transmission"),
                    "Number of Previous Owners": column_data.get("ownerNo"),
                    "Ownership Details": column_data.get("owner"),
                    "OEM": column_data.get("oem"),
                    "Car Model": column_data.get("model"),
                    "Year of Manufacture": column_data.get("modelYear"),
                    "Central Variant ID": column_data.get("centralVariantId"),
                    "Variant Name": column_data.get("variantName"),
                    "Price": column_data.get("price"),
                     })
            
            elif column == 'new_car_overview':
                for item in column_data.get('top', []):
                    car_info[item.get('key')] = item.get('value')
                    
            elif column == 'new_car_feature':
             # Check each feature category
                for category in column_data.get('data', []):
                    category_name = category.get('subHeading', '')
                    # Extract feature list for categories (like Comfort, Exterior)
                    features = [item.get('value') for item in category.get('list', [])]
                    car_info[category_name] = ', '.join(features)
            elif column == 'new_car_specs':
             # Extract the 'top' section
                for item in column_data.get('top', []):
                    car_info[item.get('key')] = item.get('value')
                
                # Process the 'data' section directly
                for category in column_data.get('data', []):
                    for spec in category.get('list', []):
                        car_info[spec.get('key')] = spec.get('value')  # Directly add key-value pairs

                
                flattened_car_data.append(car_info)

# Convert the flattened data to a DataFrame
kolkatacars_df = pd.DataFrame(flattened_car_data)

# Display the final DataFrame
print(kolkatacars_df)
```

ADD THE COLUMN CITY KOLKATA

```python
# Add a City column with the same value for all rows
kolkatacars_df['City'] = 'Kolkata'
```

```python
#import numpy as np
import pandas as pd

# Convert the 'price' column to a Series for easier calculations
price_series = pd.Series(carsdf['Price'])

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = price_series.quantile(0.25)
Q3 = price_series.quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out values outside of these bounds
# The corrected assignment to the DataFrame
carsdf['filtered_price'] = price_series[(price_series >= lower_bound) & (price_series <= upper_bound)]
```

```python
import numpy as np
import pandas as pd

# List of numerical columns (this should be the list of columns that you want to scale)
numerical_cols = [
    'Kilometers Driven', 'Number of Previous Owners', 'Engine Displacement', 'Mileage', 'Torque', 'Car Age', 'Max_Power'
]

# Start with the original DataFrame
cleaned_numerical = carsdf.copy()

# Step 1: Create a mask for the rows that will be kept after outlier removal
for column in numerical_cols:
    # Calculate the IQR for each numerical column
    Q1 = cleaned_numerical[column].quantile(0.25)
    Q3 = cleaned_numerical[column].quantile(0.75)
    IQR = Q3 - Q1

    # Calculate the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Step 2: Apply the condition for filtering based on the IQR method
    mask = (cleaned_numerical[column] >= lower_bound) & (cleaned_numerical[column] <= upper_bound)

    # Apply the mask to filter out outliers
    cleaned_numerical = cleaned_numerical[mask]

# After outlier removal, we are left only with the numerical columns
cleaned_numerical = cleaned_numerical[numerical_cols]

# Check the shape of the DataFrame after outlier removal (only for numerical columns)
print(cleaned_numerical.shape)
```
